```python
from pathlib import Path
import csv

budget_csv = Path("../Desktop/pybank.csv")

total_months = 0
pnl = []
pnl_dates = []
pnl_sum = 0
biggest_increase = 0
biggest_decrease = 0
sum = 0

with open(budget_csv, 'r') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter = ',')
    header = next(csv_reader)
    for row in csv_reader:
        
        total_months += 1
        pnl.append(int(row[1]))
        pnl_sum += int(row[1])
        pnl_dates.append(row[0])
       
pnl_change = []

for i in range(1, len(pnl)):
    x = pnl[i] - pnl[i - 1]
    pnl_change.append(int(x))
    

for i in range(0, len(pnl_change)):
    sum += pnl_change[i]
    average_change_profitloss = round((sum / (len(pnl_change))), 2)

for p_l in pnl:
    
    if biggest_decrease == 0:
        biggest_increase == p_l
        biggest_decrease == p_l
    if p_l > biggest_increase:
        biggest_increase = p_l
    elif p_l < biggest_decrease:
        biggest_decrease = p_l
        

biggest_increase_index = pnl.index(biggest_increase)
biggest_decrease_index = pnl.index(biggest_decrease)

biggest_increase_date = pnl_dates[biggest_increase_index]
biggest_decrease_date = pnl_dates[biggest_decrease_index]


output_path = Path('Financial_Analysis.txt')
with open(output_path, 'w') as file:
    file.write(f"Financial Analysis \n")
    file.write(f"----------------------------------------------------- \n")
    file.write(f"Total Months: {total_months} \n")
    file.write(f"Total: ${pnl_sum} \n")
    file.write(f"Average Change: ${average_change_profitloss}\n")
    file.write(f"Greatest Increase in Profits: {biggest_increase_date} (${biggest_increase}) \n")
    file.write(f"Greatest Decrease in Profits: {biggest_decrease_date} (${biggest_decrease}) \n")
```
